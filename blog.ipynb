```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
#to access the dataset to be used in this analysis, pandas is used to import it to the editor
#this is contained in a variable named student info
student_info = pd.read_csv("student_info.csv")
student_info.head() #.head() is a method used for the display of the first ten details
#student_info.isnull().sum() #.isnull.sum() is to check for missing values
#there is no missing value in this dataset
#.describe() method is used to provide the statistical summary of the numerical values in the dataset.
student_info.describe()
#this marks the end of the first phase of our EDA, next move to the cleaning phase
#this data has been rid of every data impunities such as missing values or outliers
#i will group my categorical data into accessible groups and store it in a variable named categorical info
categorical_info = student_info.select_dtypes(exclude=['int64','float64'])
categorical_info.drop(['name','student_id'],axis=1,inplace=True)
def iterating(categorical_info):
    column1 = len(categorical_info.columns)
    fig,axes = plt.subplots(nrows = (column1//3) ,ncols=3,figsize=(18,4 * (column1//3)))
    fig.tight_layout(pad = 5.0)
    axes = axes.flatten()
    for k,v in enumerate(categorical_info.columns):
       sns.countplot(x = v,data = categorical_info,ax=axes[k])
       axes[k].set_title(f"bar chart for {v}")
    for l in range(k + 1, len(axes)):
        fig.delaxes(axes[l])
        plt.show()

iterating(categorical_info)
#i will group my quanitative data into accessible groups and store it in a variable named quantitative info
quantitative_info = student_info.select_dtypes(include = ['int64','float64'])
quantitative_info #reducing our columns to seven
#creating a function named iterate to do the iterative job
#quantitative_info as the parameter for this function to inform python of the kind of data it will be working with
def iterate(quantitative_info):
    #the variable 'column' is created to inform python of the specific information to be used from the dataset
    column = len(quantitative_info.columns)
    #plt.subplot is used to hold all the visuals to be made for this phase of analysis
    fig,axes = plt.subplots(nrows = (column // 2),ncols = 3, figsize=(24, 6 * (column//2)))
    #fig.tight_layout is used for adjusting the space among the diagram
    fig.tight_layout(pad = 5.0)
    axes = axes.flatten()
    #use a for loop to loop through the dataset
    for key,values, in enumerate(quantitative_info.columns):
        sns.histplot(data=quantitative_info, x=values, ax=axes[key])
        axes[key].set_title(f"histogram for {values}")
        
    #this is to eliminate any extra graph that may be formed in course of visualizing 
    for j in range (key + 1,len(axes)):
        fig.delaxes(axes[j])
    plt.show()
#i run my function
iterate(quantitative_info)
#for my bivariate analysis, i used a heatmap beacuse of its pairwise feature
#correlate variable holds the correlation of this dataset
correlate = quantitative_info.corr()
sns.heatmap(correlate, cmap = 'coolwarm')
```
